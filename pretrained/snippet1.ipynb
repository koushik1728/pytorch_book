```python
from torchvision import models

dir(models)
```

```python
alexnet = models.AlexNet() # models weights and biases are random numbers, not pretrained values.
						# just using the architecture of the pretrained network with untrained weights

alexnet
```

```python
resnet = models.resnet101(pretrained=True)
```

```python
resnet
```

```python
from torchvision import transforms

preprocess = transforms.Compose([
				transforms.Resize(256),
				transforms.CenterCrop(224),
				transforms.ToTensor(),
				transforms.Normalize(
					mean=[0.485, 0.456, 0.406],
					std=[0.229, 0.224, 0.225]
				)
			])
```

```python
from PIL import Image
bobby = Image.open("./data/p1ch2/bobby.jpg")
bobby.show()
```

```python
import torch
bobby_cleaned = preprocess(bobby)
batch_t = torch.unsqueeze(bobby_cleaned, 0)
```

```python
resnet.eval()

output = resnet(batch_t)

output.size()
```

#### opening a text file

```python
with open("../dlwpt-code/data/p1ch2/imagenet_classes.txt") as file:
	labels = [line.strip() for line in file.readlines()]

labels[:5]
```

```python
output_class_index = int(output.argmax())
print(f"output class probs: {output[0, output.argmax()*0]}, output class label: {labels[output_class_index]}")
```

#### Correct way of doing this

```python
max_value, index = torch.max(output, 1)
print(max_value, index)
```

```python
percentages = torch.nn.functional.softmax(output, dim=1)[0]*100

print(percentages[index.item()].item(), labels[index.item()])
```

```python
# Second best and third best
ranking_chart_size = 10


_, indices = torch.sort(output, descending=True)
print(f"{indices.size()}\n")


top_ranks = [(labels[idx], percentages[idx].item()) for idx in indices[0][:ranking_chart_size]]
for element in top_ranks:
	print(element)
```
