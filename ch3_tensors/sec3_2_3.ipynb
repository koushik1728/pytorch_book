### 3.3

```python
some_list = torch.tensor([i for i in range(10)])

print(some_list[1:4]) # [1] inclusive (included), [4] ignored - [1], [4] means some_list[1], some_list[4]
print(some_list[1:])
print(some_list[:4])
print(some_list[:-1]) # [-1] ignored, until the element before print(some_list[-1]
print(some_list[1:4:2]) # [1] and [4] both inclusive, = [1], [3]
```

```python
# Adds a dim before dim 0 -  like some_list.unsqueeze()



print(some_list.size())
print(some_list[None].size())
```
